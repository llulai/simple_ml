```python
from scipy.optimize import minimize
import numpy as np
```

```python
def get_mse(features, y):
    def fun(coef):
        nonlocal features
        nonlocal y
        return np.sum(np.square(y - np.dot(features, coef)))
    return fun

def add_ones(self, X):
    return np.concatenate([np.ones((100, 1)), X], axis=1)


class LinearRegression:
    def __init__(self):
        self.coef_ = []
        self.intercept = None
        
    def fit(self, X, y, fit_intercept=True):
        if fit_intercept:
            features = add_ones(X)
        else:
            features = X
            
        initial_guess = np.ones(features.shape[-1])
        
        opt = minimize(get_mse(features, y), initial_guess, method='SLSQP')
        
        if fit_intercept:
            self.intercept = opt.x[0]
            self.coef_ = opt.x[1:]
        else:
            self.coef_ = opt.x
            
    def predict(self, X):
        if self.intercept is not None:
            return np.dot(X, self.coef_)
        return np.dot(add_ones(X), self.intercept + self.coef_)
```

```python
X = np.random.normal(10, 5, (100, 5))

b = np.random.normal(3, 10, 5)

u = np.random.normal(0, 1, 100)

a = 3

y = a + np.dot(X, b) + u
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
reg.predict(X).astype(float)
```
